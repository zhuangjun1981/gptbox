```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import copy
```

```python
# url = "https://www.space.com/spacex-starship-launch-debris-terrifying"
url = "https://www.space.com/spacex-starship-damage-starbase-launch-pad"
```

```python
html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")
```

```python
soup2 = copy.deepcopy(soup)

# Find all unwanted elements such as ads, links, buttons etc.
unwanted_tags = ['aside', 'script', 'style', 'footer']
for tag in unwanted_tags:
    for element in soup2.find_all(tag):
        element.decompose()
        
for element in soup2.find_all('div'):
    if element.get('class') is not None:
        if element.get('class')[0] in ['more-about__cardBody']:
            element.decompose()


stuff = {}

metas = soup2.find('head').find_all('meta')
# [print(meta) for meta in metas]

for meta in metas:
    if meta.get('property') == 'og:title':
        stuff['title'] = meta.get('content').strip()
        
    if meta.get('property') == 'og:description':
        stuff['description'] = meta.get('content').strip()
        
    if meta.get('property') == 'og:url':
        stuff['url'] = meta.get('content').strip()
        
ps = soup2.find_all('p')
# [print(p) for p in ps]

article = []
pi = 0
while pi < len(ps):
    
    p = ps[pi]
    ptxt = p.get_text().strip()
    
    if p.get('class') is not None:
        if p.get('class')[0] in ['affiliate-disclaimer__copy', 'strapline',
                                 'newsletter-form__strapline', 'article__byline']:
            pi += 1
            continue
        
    if p.find('strong') is not None: # remove the section starts with "related"
        pi += 1
        continue
        
    if p.get('dir') == 'ltr': # remove twitter insert
        pi += 1
        continue
    
    if ptxt.startswith('—  '):
        pi += 1
        continue
        
    if ptxt.startswith('— '):
        pi += 1
        continue
        
    if p.find('em') is not None:
        pi += 1
        continue
            
    article.append(ptxt)
    pi += 1
    
article = '\n'.join([a for a in article if a])
article = article.replace(" (opens in new tab)", "")

print('\n\n')
for key, value in stuff.items():
    print(f'{key}: {value}')
    
print('\n\n')
print(article)
```
